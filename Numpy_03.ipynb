### More Numpy Functions

```python
import numpy as np
```

### Sorting Arrays

```python
a = np.random.randint(1,100,15)
b = np.random.randint(1,100,15).reshape(3,5)
print(a,'\n\n', b)
```

```python
np.sort(a)
```

```python
np.sort(b) # By default row-wise sorting
```

```python
np.sort(b, axis=0) # Column-wise sorting
```

```python
np.sort(a)[::-1] # Descending sorting
```

```python
np.sort(b, kind='mergesort')
```

```python
# Concatenate
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print()
print(b)
```

```python
np.concatenate((a,b)) # Similar to vstack
```

```python
np.concatenate((a,b), axis=1) # Similar to hstack
```

```python
# Unique
e = np.array([1,1,1,2,2,3,4,5,5,5,6])
np.unique(e)
```
