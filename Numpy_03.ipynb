### More Numpy Functions

```python
import numpy as np
```

### Sorting Arrays

```python
a = np.random.randint(1,100,15)
b = np.random.randint(1,100,15).reshape(3,5)
print(a,'\n\n', b)
```

```python
np.sort(a)
```

```python
np.sort(b) # By default row-wise sorting
```

```python
np.sort(b, axis=0) # Column-wise sorting
```

```python
np.sort(a)[::-1] # Descending sorting
```

```python
np.sort(b, kind='mergesort')
```

### Set functions

```python
m = np.arange(1,6)
n = np.arange(3,8)

print(m,'\n\n',n)
```

```python
# Union
np.union1d(m,n)
```

```python
# Intersection
np.intersect1d(m,n)
```

```python
# Set difference
np.setdiff1d(m,n)
```

```python
np.setdiff1d(n,m)
```

```python
# xor
np.setxor1d(m,n)
```

```python
# Clip
print(a)
np.clip(a, a_max=80, a_min=20)
```
