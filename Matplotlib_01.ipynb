### Types of Data
- Numerical Data
    - Data consisting only number
- Categorical Data
    - Data consisting categories values
    - Eg. A column of gender in a dataset. There can only be some values in it but can be same value of multiple rows. Categories can be created on that basis

### Types of Analysis
- Univariate Analysis
    - Doing analysis based on 1 column
- Bivariate Analysis
    - Doing analysis based on 2 columns
- Multivariate Analysis
    - Doing analysis based on more than 2 columns

### Scatter Plot
- Bivariate Analysis
- Numerical - Numerical
- Use case - Finding correlation

```python
# plotting simple scatter plot
x = np.linspace(-10,10,50)
y = 10*x+3 + np.random.randint(0,300,50)

plt.scatter(x,y)
```

```python
# plotting on pandas dataset
df = pd.read_csv('Matplotlib_Dataset\\batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'], marker='x')

plt.title('Avg and SR analysis of top 50 Batsman')
plt.xlabel('Avg')
plt.ylabel('SR')
```

```python
# size
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10) # smaller dots represents smaller values, larger dots represents larger values
```

```python
# scatterplot using plt.plot
plt.plot(tips['total_bill'], tips['tip'], 'o') # faster technique but can't manipulate size and color
```

### Bar Chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['Red', 'Blue', 'green', 'yellow', 'pink']

plt.bar(colors, children, color= ['Red', 'Blue', 'green', 'yellow', 'pink']) # Vertical bar chart
```

```python
# horizontal bar chart
plt.barh(colors, children, color= ['Red', 'Blue', 'green', 'yellow', 'pink'])
```

```python
# multiple bar chart
df = pd.read_csv('Matplotlib_Dataset\\batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2016'], label='2016')
plt.bar(df['batsman'], df['2015'], label='2015')

plt.legend()
```

- 2016 data graph is below 2015 data graph and you can't see values side by side of both Year.

```python
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width=0.3, label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.3, label='2016')
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width=0.3, label='2017')

plt.legend()
plt.xticks(np.arange(df.shape[0]), df['batsman']) # change x-axis labels
plt.show()
```

- A little too complex in matplotlib, easier in seaborn.

```python
children = [10,20,40,10,30]
colors = ['Redddddddddddddddddddd', 'Blueeeeeeeeeeeeeeeeeeeeeee', 'greennnnnnnnnnnnnnnnn', 'yellowwwwwwwwwwwwwwwww', 'pinkkkkkkkkkkkk']

plt.bar(colors, children)
plt.xticks(rotation=90)
plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'], df['2017'], bottom=df['2016']+df['2015'], label='2017')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label='2016')
plt.bar(df['batsman'], df['2015'], label='2015')

plt.legend()
```

### Pie Chart
- Univariate/ Bivariate Analysis
- Categorical vs Numerical
- Use case - To find contribution on a standard scale

```python
# simple data
df = np.random.randint(1,100,5)
subjects = ['eng','sci','maths','hindi','sst']
df
```

```python
plt.pie(df, labels=subjects)
plt.show()
```

```python
# dataset
df = pd.read_csv('Matplotlib_Dataset\\gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', startangle=45, explode=[0.1,0,0,0,0,0], shadow=True)
plt.show()
```
