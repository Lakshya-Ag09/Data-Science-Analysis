### Types of Data
- Numerical Data
    - Data consisting only number
- Categorical Data
    - Data consisting categories values
    - Eg. A column of gender in a dataset. There can only be some values in it but can be same value of multiple rows. Categories can be created on that basis

### Types of Analysis
- Univariate Analysis
    - Doing analysis based on 1 column
- Bivariate Analysis
    - Doing analysis based on 2 columns
- Multivariate Analysis
    - Doing analysis based on more than 2 columns

```python
# improting library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2D Plot
- Bivariate Analysis
- Categorical - Numerical and Numerical - Numerical
- Use case - Time series data

```python
# plotting a simple graph
price = [48000,54000,56000,49000,47000,45000]
year = np.arange(2015,2021)

plt.plot(year, price) # plt.plot(x axis, y axis)
```

```python
# plotting graph from pandas dataframe
batsman = pd.read_csv('Matplotlib_Dataset\\sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors and line(width and style) and marker size
plt.plot(batsman['index'], batsman['V Kohli'], color='aqua', linestyle='dashed', linewidth=3, marker='d')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted', linewidth=2, marker='+')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='aqua', linestyle='dashdot', linewidth=3, marker='.', markersize=20)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dashdot', linewidth=2, marker='<')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend
plt.plot(batsman['index'], batsman['V Kohli'], label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend() # default location is best, which means it will be added on best location on graph
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='lower right')
```
