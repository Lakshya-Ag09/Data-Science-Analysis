### **Seaborn**
- Simpler to use for visualization
- Better aesthetics
- More graphs included

**Types of Functions:**
- Figure level
- Axis level

**Analogy:**
- **Figure:** Think of it like container where the graphs are shown
- **Axis:** Think of it like graphs in that container
- Graphs can be plot on figure level or axis level

**Classification of Graphs:**
1. **Relational:** Plotting relationship between 2 quantities
    - Scatter plot
    - Line plot
2. **Distribution:** Plotting on single quatity
    - Histogram
    - KDE plot
    - Rug plot
3. **Categorical:** Plotting on categorical data
    - Bar plot
    - Count plot
    - Box plot
    - Voilon plot
    - Swarm plot
4. **Matrix:** Used to plot on grid of data
    - Heatmap
    - Cluster map
5. **Regression:** Used for regression model 
    - lmplot
    - Reg plot
    - Resid plot
6. **Multiplots:** Plotting multiple related plots
    - Joint plot
    - Pair plot

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**1. Relational Plot**
- to see the statistical relation between 2 or more variable
- Bivariate Analysis
- scatter plot, line plot

```python
# scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# relplot -> figure level function (generally in square shape)
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
# legend is out of graph inside the figure
```

```python
# line plot -> axis level function
gap = px.data.gapminder()
df = gap[gap['country'] == 'India']
df.head()
```

```python
sns.lineplot(data=df, x='year', y='lifeExp')
```

```python
# using relpot
sns.relplot(data=df, x='year', y='lifeExp', kind='line')
```

```python
df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
df.head()
```

```python
sns.relplot(kind='line', data=df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
# facet plot -> figure level function -> work with relplot but not with scatter plot or line plot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='smoker', row='sex')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=2)
```

**2. Distribution Plots**
- Univariate Analysis
- use to find out distribution
- range of observation
- Central tendency (mean, median)
- Is the data bimodal? (means multiple peaks)
- Are there outliers?
- histplot, kdeplot, rugplot

```python
# histplot -> axis level function
tips.head()
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
# displot -> figure level function
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

```python
sns.displot(data=tips, x='day', kind='hist')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(kind='hist', data=titanic, x='age', element='step', hue='sex')
```

```python
# facet plot
sns.displot(kind='hist', data=titanic, x='age', element='step', col='sex')
```

```python
# kde(kernal density estimation) plot -> axis level function
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(kind='kde', data=tips, x='total_bill', hue='sex', fill=True, height=5, aspect=2)
```

```python
# rugplot -> axis level function
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
# used to complement other plots by showing additional information around the ticks
```

```python
# Bivariate histogram

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(kind='hist', data=tips, x='total_bill', y='tip')
```

```python
# Bivariate kdeplot

# sns.kdetplot(data=tips, x='total_bill', y='tip')
sns.displot(kind='kde', data=tips, x='total_bill', y='tip')
```

**3. Matrix Plot**
- heatmap, clustermap

```python
# heatmap -> axis level function (no figure level function for this plot)
df = gap.pivot(index='country', columns='year', values='lifeExp')
df.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data=df)
```

```python
df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
df.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data=df, annot=True, linewidths=0.5, cmap='viridis')
```

```python
# clustermap -> figure level function
iris = px.data.iris()
iris.head()
```

```python
sns.clustermap(iris.iloc[:,:4])
```

**4. Categorical Plots**
- Categorical Scatter Plots
    - Strip plot
    - Swarm plot
- Categorical Distribution Plots
    - Box plot
    - Voilin plot
- Categorical Estimate Plots -> for central tendency
    - Bar plot
    - Point plot
    - Count plot

- **Categorical Scatter Plots**

```python
# strip plot -> axis level function
tips.head()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
# catplot -> figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=0.2, hue='sex')
```

```python
# swarmplot -> axis level plot
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='sex')
```

- **Categorical Distribution Plot**

```python
# box plot -> axis level function
sns.boxplot(data=tips, x='sex', y='total_bill')
```

```python
sns.catplot(kind='box', data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.catplot(kind='box', data=tips,y='total_bill')
```

```python
# voilon plot (boxplot + kdeplot) -> axis level function
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True)
```

- **Categorical Estimate Plot**

```python
# barplot -> axis level function
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='bar', errorbar=None, hue='smoker')
```

```python
# estimator
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.std)
```

```python
# point plot -> axis level function
sns.pointplot(data=tips, x='sex', y='total_bill') # connecting points of bar plot
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill', color='orange')
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
sns.pointplot(data=tips, x='sex', y='total_bill', errorbar=None)
```

```python
sns.catplot(data=tips, x='sex', y='total_bill', kind='point')
```

```python
# count plot -> axis level plot
sns.countplot(data=tips, x='sex', hue='day')
```

```python
sns.catplot(data=tips, x='sex', hue='day', kind='count')
```

```python
# facet plot
sns.catplot(data=tips, x='sex', y='total_bill', col='smoker', kind='box', row='time')
```

**5. Regression Plots**
- regplot, lmplot, residplot

```python
# regplot -> axis level function
sns.regplot(data=tips, x='total_bill', y='tip') # cannot add hue
```

```python
# lmplot -> figure level function
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex') # hue can be added
```

```python
# residplot ->
sns.residplot(data=tips, x='total_bill', y='tip')
```
