**Why separate objects to handle data and time when python already has datetime functionality?**
- Syntax wise datetime is very convenient, but the performance takes a hit while working with huge data. Similar as List vs Numpy Array.
- The weaknesses of Python's datetime format inspired the Numpy team to add a set of native time series data type to Numpy.
- The datetime64 dtype encodes dates as 64-bit int, and thus allows arrays of dates to be represented very compactly.

```python
date = np.array('2022-07-03', dtype=np.datetime64)
date
```

```python
date + np.arange(12)
```

- Because of the uniform type in Numpy datetime64 arrays, this type of operations can be accomplished much more quickly than if we were working directly with Python's datetime objects, especially as arrays get large.
- Pandas Timestamps object combines the ease-of-use of python datetime with the efficient storage and vectorized interface  of numpy.datetime64
- From a group of these Timestamp objects, Pandas can constuct a DatetimeIndex that can be used to index data in a Series or Dataframe.

### DatetimeIndex object
- A collection of pandas timestamp.

```python
# creating using strings
d = pd.DatetimeIndex(['2022/2/2','2021/1/1','2090/3/23'])
d
```

```python
d[0]
```

```python
type(d)
```

```python
# creating using python datetime object
pd.DatetimeIndex([dt.datetime(2022,1,1),dt.datetime(2022,3,31),dt.datetime(2122,2,2)])
```

```python
# creating using timestamps
dt_index = pd.DatetimeIndex([pd.Timestamp(2022,1,1),pd.Timestamp(2022,3,31),pd.Timestamp(2122,2,2)])
dt_index
```

```python
# using datetimeindex as series index
pd.Series([1,2,3], index=dt_index)
```

```python
pd.Series(dt_index)
```

### date_range function

```python
# generates daily dates in a given range
pd.date_range(start='1 jan 2022', end='1 mar 2022', freq='D') # D -> daily
```

```python
pd.date_range(start='1 jan 2022', end='1 mar 2022', freq='2D') # 2D -> skips 1 day
```

```python
pd.date_range(start='1 jan 2022', end='1 mar 2022', freq='B') # B -> Business days
```

```python
pd.date_range(start='1 jan 2022', end='1 mar 2022', freq='W') # W -> 1 day of week, by default sunday
```

```python
pd.date_range(start='1 jan 2022', end='1 mar 2022', freq='W-THU')
```

```python
pd.date_range(start='1 jan 2022', end='1 mar 2022', freq='h') # h -> Hourly data
```

### to_datetime function
- Converts an existing object to pandas timestamp/datetimeindex object.

```python
s = pd.Series(['2022/1/1','2000/1/1','2033/2/4'])
s
```

```python
pd.to_datetime(s)
```

```python
# datetime accessor
pd.to_datetime(s).dt.day_name()
```

```python
s = pd.Series(['2022/1/1','2000/1/1','2033/200/4']) # month is 200 which is invalid
s
```

```python
pd.to_datetime(s, errors='coerce') # to avoid throw error, return NaT(Not a Time)
```
