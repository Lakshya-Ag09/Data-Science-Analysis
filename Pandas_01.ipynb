### **Pandas Series**
Pandas Series and it's operations

---

Pandas used for :
- Data Analysis
- Data Manipulation

---
Pandas has 2 objects :
- Series
- Dataframe

Series: 1-D array of data of any type. Just like a column in a table

```python
# Importing Pandas
import numpy as np
import pandas as pd
```

### Initializing Series

```python
# List of Strings values
country = ['India', 'Pakistan', 'USA', 'Nepal', 'Bangladesh']
pd.Series(country)
```

- As you can see here, series contains **Index** and **Values**

```python
# List of  Numeric values
runs = np.random.randint(1,100,5)
pd.Series(runs)
```

```python
# Custom Index
marks = np.random.randint(50,100,4)
subjects = ['Maths', 'English', 'Physics', 'Chemistry']

pd.Series(marks, index=subjects)
```

```python
# Naming a series
pd.Series(marks, index=subjects, name='Scorecard')
```

```python
# Initializing series from dict
marks = {
    'Bio':87,
    'English':91,
    'Physics':50,
    'Chemistry':67
}

pd.Series(marks)
```

```python
# List of different data types
li = ['Lakshya',40,29,'Akshat',48]
pd.Series(li)
```

- Here, object dtype means **string values** and int values are treated as strings

### Indexing and Slicing

```python
# Integer indexing
x = pd.Series(np.random.randint(1,100,10))
print(x)
x[4]
```

```python
x[-1] # Negative indexing don't work here
```

```python
print(lead)
print(lead['Battalion 609'])
print(lead[1]) # This doesn't work here
```

```python
# Slicing
runs[5:11]
```

```python
runs[-5:] # Negative slicing works
```

```python
lead[-5:] # Integer slicing works on custom indices
```

```python
# Fancy Indexing
runs[[1,4,5,8]] # Any random index
```

### Series with Python functionality

```python
# len/type/dir/sorted/max/min
print(len(subs))
print(type(subs))
print(dir(subs))
print(sorted(subs))
print(max(subs))
print(min(subs))
```

```python
# type conversion
print(ms)
list(ms)
```

```python
dict(ms)
```

```python
# membership operator
print(lead)
'Why Cheat India' in lead
```

```python
print('Vicky Kaushal' in lead) # Works on index
print('Vicky Kaushal' in lead.values) # Works on values
```

```python
# looping
for i in lead:
    print(i) # Prints all the values
```

```python
for i in lead.index:
    print(i) # Print all indices
```

```python
# Arithmetic operators
print(ms)
print(100-ms)
print(100+ms)
```

```python
# Relational operators
print(runs.head())
print(runs.head()>50)
print(runs[runs>50].head()) # Boolean Masking
```

```python
# Value counts
print(lead)
num = lead.value_counts() # Gives frequency of values
```

```python
# Boolean masking
print(num[num>20])
print(num[num>20].size)
```
