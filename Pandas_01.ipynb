### **Pandas Series**
Pandas Series and it's operations

---

Pandas used for :
- Data Analysis
- Data Manipulation

---
Pandas has 2 objects :
- Series
- Dataframe

Series: 1-D array of data of any type. Just like a column in a table

```python
# Importing Pandas
import numpy as np
import pandas as pd
```

### Initializing Series

```python
# List of Strings values
country = ['India', 'Pakistan', 'USA', 'Nepal', 'Bangladesh']
pd.Series(country)
```

- As you can see here, series contains **Index** and **Values**

```python
# List of  Numeric values
runs = np.random.randint(1,100,5)
pd.Series(runs)
```

```python
# Custom Index
marks = np.random.randint(50,100,4)
subjects = ['Maths', 'English', 'Physics', 'Chemistry']

pd.Series(marks, index=subjects)
```

```python
# Naming a series
pd.Series(marks, index=subjects, name='Scorecard')
```

```python
# Initializing series from dict
marks = {
    'Bio':87,
    'English':91,
    'Physics':50,
    'Chemistry':67
}

pd.Series(marks)
```

```python
# List of different data types
li = ['Lakshya',40,29,'Akshat',48]
pd.Series(li)
```

- Here, object dtype means **string values** and int values are treated as strings

### Series Attributes

```python
ms = pd.Series(marks, name='Scorecard')
ms
```

```python
# size
ms.size # Number of elements in series
```

```python
# dtype
ms.dtype # Type of element
```

```python
# name
ms.name # Name of series
```

```python
# is_unique
ms.is_unique # Tells if every item is unique or not
```

```python
pd.Series([1,1,1,3,5,6,8,2]).is_unique
```

```python
# index
ms.index # Gives index values
```

```python
pd.Series([1,1,1,3,5,6,8,2]).index
```

```python
# values
ms.values # Gives values of series
```
