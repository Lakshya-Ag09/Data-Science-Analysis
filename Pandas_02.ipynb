### **Pandas Dataframe**
Dataframe: 2-D structure of data consisting of rows and column, basically tables in Pandas are called Dataframe
- Every individual column/row of Dataframe is an individual Series

```python
import numpy as np
import pandas as pd
```

### Creating Dataframe

```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data, columns=['IQ','Marks','Package'])
```

```python
# using dicts
student_dict = {
    'IQ':[100,90,120,80],
    'Marks':[80,70,100,50],
    'Package':[10,7,14,2]
}

student = pd.DataFrame(student_dict)
```

### Filtering a dataframe

```python
ipl.head(2)
```

```python
# find all the final winners
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam']]

ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']] # one liner
```

```python
# how many superover finishes have occured
ipl[ipl['SuperOver'] == 'Y']['ID'].count()
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches has csk won in kolkata
(ipl[ipl['City'] == 'Kolkata']['WinningTeam'] == 'Chennai Super Kings').sum()

ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0] # other way
```

```python
# action movies with rating higher than 7.5
movies.sample(2)
```

```python
mask1 = movies['genres'].str.split('|').apply(lambda x: 'Action' in x)
mask2 = movies['IMDB_rating'] > 7.5
movies[mask1 & mask2]['original_title']
```
