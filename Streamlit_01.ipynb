### **Streamlit**
- Used to make web apps with basic and already made customizations
- No need to understand frontend backend basics for web app
- Frontend is handled by react and backend is handled by python
- No customizations can be done

To run streamlit site, you don't run the file directly but:
- Go to terminal 
- Type: streamlit run [filename]
- Streamlit only runs file with .py extension and no other extension

```python
import streamlit as st
import pandas as pd
import time
```

**Text Utility**

```python
# adding title
st.title('Startup Dashboard')
```

```python
# adding header
st.header('I am learning Streamlit')
```

```python
# adding subheader
st.subheader('And I am loving it!')
```

```python
# adding normal text
st.write('This is a normal text')
```

```python
# adding markdown
st.markdown("""
### My favourite movie
- Race 3
- Humshakal
- Housefull
""")
```

```python
# adding code
st.code('''
def name(text):
    return text
x = name('Lakshya')
''')
```

```python
# adding latex
st.latex('''
sin^2x + cos^2x = 1 
''')
```

**Display Elements**

```python
df = pd.DataFrame({
    'name':['Nitish', 'Lakshya', 'Harsh'],
    'marks':[50,60,70],
    'package':[10,20,30]
})
```

```python
# adding dataframe
st.dataframe(df)
```

```python
# adding json
st.json({
    'name':['Nitish', 'Lakshya', 'Harsh'],
    'marks':[50,60,70],
    'package':[10,20,30]
})
```
