### **Plotly**
Data Visualization Library

**Advantages:**
- Multi language support (Python, JS, Julia)
- Lot's of graphs
- Interactive plots
- Beautiful plots

**Three main components:**
- Graph Objects (go)
- Plotly Express
- Dash

```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import pandas as pd
```

### Working with Plotly go

```python
# importing dataset
tips = px.data.tips()
iris = px.data.iris()
gap = px.data.gapminder()
```

- Scatter plot ->

```python
# scatter plot using plotly go
df = gap[gap['year'] == 2007]
df.head()
```

```python
trace1 = go.Scatter(x=df['lifeExp'], y=df['gdpPercap'], mode='markers')
trace2 = go.Scatter(x=[0,10,40], y=[0,9000,30000], mode='lines')
layout = go.Layout(title='Life Exp Vs GDP per Capita for 2007', xaxis={'title':'Life Exp', 'color':'aqua'}, yaxis={'title':'GDP'})

data = [trace1, trace2]

fig = go.Figure(data, layout)
fig.show()
```

```python
# scatter plot using plotly express
px.scatter(df, x='lifeExp', y='gdpPercap', color='continent', size='pop', size_max=50, hover_name='country')
```

```python
# plotting animation
px.scatter(gap, x='lifeExp', y='gdpPercap', color='continent', size='pop',
            size_max=50, hover_name='country', animation_frame='year',
            animation_group='country', range_x=[30,90], range_y=[-10000,51000])
```

- 3D plots ->

```python
# scatter plot
df = gap[gap['year'] == 2007]
df.head()
```

```python
px.scatter_3d(df, x='lifeExp', y='pop', z='gdpPercap', log_y=True)
```

```python
# surface plot
# cannot be created using plotly express but by graph objects
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2

trace1 = go.Surface(x=x, y=y, z=z)
data = [trace1]
layout = go.Layout(title='3D Surface Plot')

fig = go.Figure(data, layout)
fig.show()
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)
z = np.sin(xx) + np.cos(yy)

trace1 = go.Surface(x=x, y=y, z=z)
data = [trace1]
layout = go.Layout(title='3D Surface Plot')

fig = go.Figure(data, layout)
fig.show()
```

- Contour plot ->

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2

trace1 = go.Contour(x=x, y=y, z=z)
data = [trace1]
layout = go.Layout(title='Contour Plot')

fig = go.Figure(data, layout)
fig.show()
```

- Subplots ->

```python
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(rows=1, cols=2)
```

```python
fig.add_trace(
    go.Scatter(x=[1,9,4], y=[2,10,3], mode='markers'),
    row=1,
    col=1
)
fig.add_trace(
    go.Histogram(x=[1,9,4,22,124,34,5,67,45,78,54]),
    row=1,
    col=2
)

fig.update_layout(title='Subplot Demo')
```
