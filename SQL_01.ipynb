### **Databases**
A Database is a shared collection of logically related data and description of these data, designed to meet the information needs of an organization.
- **Data Storage:** A database is used to store large amounts of structured data, making it easily accessible, searchable, and retrievable.
- **Data Analysis:** A database can be used to perform complex data analysis, generate reports, and provide insights into the data.
- **Record Keeping:** A database is often used to keep track of important records, such as financial transactions, customer information, and inventory levels.
- **Web Applications:** Databases are an essential component of many web applications, providing dynamic content and user management.

### Types of Databases
- **Relational Databases**
    - Also known as RDBMS / SQL databases, these databases use a relational model to organize data into tables with rows and columns.
- **NoSQL Databases**
    - These databases are designed to handle large amounts of unstructured or semi-structured data, such as documents, images, or videos. (MongoDB)
- **Column Databases**
    - These databases store data in columns rather than rows, making them well-suited for data warehousing and analytical applications. (Amazon Redshift, Google BigQuery)
    - Think like there is a new table with only index and single column of a table contain multiple rows and columns thus giving each table for each column.
- **Graph Databases**
    - These databases are used to store and query graph-structured data, such as social network connections or recommendation systems. (Neo4j, Amazon Neptune)
    - Think of it like data stored in tree like or graph like structure.
- **Key-value Databases**
    - These databasses store data as a collection of keys and values, making them well-suited for caching and simple data storage needs. (Redis, Amazon DynamoDB)
    - Think of it like data stored in Key-Value pairs like dictionary in Python.

### DBMS (DataBase Management System)
A DBMS is a software system that provides the interfaces and tools needed to store, organize, and manage data in a database. A DBMS acts as an intermediary between the database and the application or users that access the data stored in database.

Think of it like:
- User interacts with DBMS
- DBMS interacts with Database
- Database interacts with OS
- OS interacts with Hardware
- Then info is accessed and same goes other way around

---
### Functions of DBMS
- **Data Management:** Store, retrieve, and modify data
- **Integrity:** Maintain accuracy of data
- **Concurrency:** Simultaneous data access for multiple users
- **Transaction:** Modification to database must be either successful or must not happen at all
- **Security:** Access to authorized users only
- **Utilities:** Data import/export, user management, backup, logging

### Database Keys
A key in a database is an attribute or set of attributes that uniquely identifies a tuple in a relation. Keys play a crucial role in ensuring the integrity and establishing relationships between tables.
- **Super Key**
    - It is a combination of columns that uniquely identifies any row within a RDBMS table
- **Candidate Key**
    - It is a minimal Super Key, meaning it has no redundant attributes. In other words, it's the smallestt set of attributes that can be used to uniquely identify tuple in the table
- **Primary Key**
    - It is a **unique identifier** for each tuple in a table. There can only be one Primary Key in a table, and it cannot contains **null values**
    - It will be great if it have **numeric values**, **small in length**, and **constant**
- **Alternate Key**
    - Candidate Keys which are not Primary Key
- **Composite Key**
    - Primary Key made up of more than 1 attributes
- Foreign Key
    - It is a Primary Key from one table that is used to establish a relationship with another table

### Cardinality of Relationships
Cardinality in database relationships refers to the number of occurences of an entity(real-world objects like cars, human, etc...). Cardinality defines the number of instances of one entity that can be associated with a single instance of the related entity.


---
### Types of Relationships
- **One-to-One Relationships**
    - One value of a relation is connected to One value of another relation
    - This needs only 1 table
    - Person -> Driving License Number
- **One-to-Many Relationships**
    - One value of a relation is connected to Many values of another relation or vice-versa
    - This needs exact 2 tables
    - Student -> College Branch
- **Many-to-Many Relationships**
    - Many values of a relation is connected to Many values of another relation and vice-versa
    - This needs exact 3 tables
    - Students -> Online Courses

### Drawbacks of Databases
- **Complex** to setup and maintain, especially for large and complex systems.
- **Cost** of setting up and maintaining can be high.
- **Scalability** issues as database grows, it can become more difficult to  manage.
- **Data Integrity** can be challenge, especially when multiple users are updating the data simultaneously.
- **Security** can be difficult as unauthorized access and protecting sensitive information can be difficult, especially with the increasing threat of cyber attacks.
- **Data Migration** can be complex and time-consuming process.
- **Flexibility** of database is often rigid, making it difficult to adapt to changing requirements or to accommodate new types of data.
