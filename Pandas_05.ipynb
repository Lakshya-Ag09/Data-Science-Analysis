### **Multi-index Series and Dataframes**
- Series -> 1D object
    - You need 1 info to fetch details.
- Dataframe -> 2D object
    - You need 2 info to fetch details.

```python
import numpy as np
import pandas as pd
```

```python
# Can we have multiple index? Let's try
index_val = [('cse', 2019),('cse', 2020),('cse', 2021),('cse', 2022),('ece', 2019),('ece', 2020),('ece', 2021),('ece', 2022)]
a = pd.Series(np.arange(1,9), index=index_val)
a
```

```python
a[('cse', 2020)]
```

**The problem?**

If we want all students of cse.

a['cse'] ---> **Error**

**The solution**

Use **Multi-index** (also known as **Hierarchical Indexing**) objects.

### Creating Multi-index objects

```python
# MultiIndex
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
multiindex.levels
```

```python
multiindex.levels[0]
```

```python
multiindex.levels[1]
```

```python
pd.MultiIndex.from_product([['cse','ece'], [2019,2020,2021,2022]]) # Cartesian product on both list
```

```python
s = pd.Series(np.arange(1,9), index=multiindex)
s
```

```python
s[('cse',2022)]
```

```python
s['cse']
```

```python
s['cse'][2020]
```

- Multi-index Series -> 2D object
    - You need 2 info to fetch details.
- Similar to Dataframe

```python
# unstack
temp = s.unstack()
temp
```

- Make Multi-index Series a Dataframe    
    - Make inner level indices, columns of Dataframe.

```python
# stack
temp.stack()
```

- Make Dataframe a Multi-index Series
    - Make columns of Dataframe, inner level indices.

### **Question**
Why we use Multi-index Series if we can use Dataframes that works similar to that and give similar results?
### **Answer**
It's not just about 1D and 2D data. There can be more dimensions of data like 3D, 4D, ... 10D ... so on. You can represent that data into 2D data directly using Multi-index objects.

```python
# Multi-index Dataframe
branch_df1 = pd.DataFrame(np.arange(1,17).reshape(8,2), index=multiindex, columns=['avg_package', 'students'])
branch_df1
```

```python
branch_df1.loc['cse']
```

```python
branch_df1['students']
```

```python
branch_df1.loc[('cse', 2020)]
```

```python
branch_df1.loc[('cse', 2020)]['students']
```

```python
# Multi-index df from columns perspective
branch_df2 = pd.DataFrame(np.arange(1,17).reshape(4,4), index=[2019,2020,2021,2022], columns=pd.MultiIndex.from_product([['Delhi','Mumbai'], ['Avg_Package','Students']]))
branch_df2
```

### Long vs Wide Data Format
- **Wide Format** is where we have a single row for every data point with multiple columns to hold the values of various attributes.
    - Think like, whenever the new data entry is to be filled then it is filled by adding a new column.
- **Long Format** is where, for each data point we have as many rows as the number of attributes and each row contains value of particular attribute for a given data point.
    - Think line, whenever the new data entry is to be filled then it is filled by adding a new row.

```python
# melt
temp = pd.DataFrame({'cse':[120], 'ece':[100], 'mech':[40]})
temp
```

```python
temp.melt() # Convert wide format to long format
```

```python
temp.melt(var_name='branch', value_name='num_students')
```

```python
temp = pd.DataFrame({
    'branch':['cse','ece','mech'],
    '2020':[100,120,90],
    '2021':[120,200,180],
    '2022':[180,90,210]
})
temp
```

```python
temp.melt()
```

```python
temp.melt(id_vars=['branch'], var_name='year', value_name='students')
```


### Pivot Table

```python
import seaborn as sns

df = sns.load_dataset('tips')
df.head()
```

```python
df.groupby('sex')['total_bill'].mean()
```

```python
df.groupby(['sex','smoker'])['total_bill'].mean().unstack()
```

```python
df.pivot_table(index='sex', columns='smoker', values='total_bill') # Brings summary of higher dimensional data
```

```python
df.pivot_table(index='sex', columns='smoker', values='total_bill', aggfunc='sum')
```

```python
df[df.select_dtypes('number').columns].pivot_table(index=df['sex'], columns=df['smoker'])['tip']
```

```python
df.pivot_table(index=['sex', 'smoker'], columns=['time', 'day'], values='total_bill')
```

```python
df.pivot_table(index='sex', columns='smoker', values='total_bill', aggfunc='sum', margins=True)
```
