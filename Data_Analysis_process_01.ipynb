- Using web scrapping to gather data

```python
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
len(company)
```

```python
name = []
rating = []
reviews = []

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div', class_='rating_text').text.strip())
    reviews.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())

d = {'name':name, 'rating':rating, 'reviews':reviews}
df = pd.DataFrame(d)
```

```python
df
```

```python
final = pd.DataFrame()

for j in range(1,11):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(f'https://www.ambitionbox.com/list-of-companies?page={j}', headers=headers).text

    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_='companyCardWrapper')

    name = []
    rating = []
    reviews = []

    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div', class_='rating_text').text.strip())
        reviews.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip())

    d = {'name':name, 'rating':rating, 'reviews':reviews}
    df = pd.DataFrame(d)
    final = pd.concat([final, df], ignore_index=True)
```

```python
final
```

```python
final.to_csv("Data Analysis Process Dataset\\companies.csv", index=False)
```
