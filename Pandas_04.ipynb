### Join Functions
- First Dataframe is left table.
- Second Dataframe is right table.
- All columns of both Dataframe will be added.
- Apply on a common column between both Dataframe.

```python
# inner join
students.merge(regs, how='inner', on='student_id')
```


- Only common values of common column will be added.

```python
# left join / join
courses.merge(regs, how='left', on='course_id')
```

- All values of left table will be added with corresponding values of right table.

```python
# right join
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Lakshya','Appar'],
    'partner':[28,26,17]
})

students = pd.concat([students, temp_df], ignore_index=True)
students
```

```python
regs.merge(students, how='right', on='student_id')
```

- All values of right table will be added with corresponding values of left table.

```python
# outer join / full join
students.merge(regs, how='outer', on='student_id').tail(10)
```

- All values of both table will be added.

```python
# self join
students.merge(students, how='inner', left_on='partner', right_on='student_id')[['name_x','name_y']]
```

- left_on tells column for left table.
- right_on tells column for right table.

### Examples

```python
# Find total revenue generated
courses.merge(regs, how='inner', on='course_id')['price'].sum()
```

```python
# Find month by month revenue
temp_df = pd.concat([nov, dec], keys=['Nov', 'Dec']).reset_index()
temp_df.merge(courses, on='course_id').groupby('level_0')['price'].sum()
```

```python
# Print the registration table
# cols -> name --- course --- price
students.merge(regs, on='student_id').merge(courses, on='course_id').drop(columns=['student_id','partner','course_id'])
```

```python
# Plot bar chart for revenue/course
regs.merge(courses, on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
# Find students who enrolled in both months
nov.merge(dec, on='student_id').merge(students, on='student_id')['name'].drop_duplicates().values
```

```python
# Find courses that got no enrollment
course_id_list = np.setdiff1d(courses['course_id'], regs['course_id'])
courses[courses['course_id'].isin(course_id_list)]
```

```python
# Find students who did no enrollment
student_id_list = np.setdiff1d(students['student_id'], regs['student_id'])
students[students['student_id'].isin(student_id_list)]
```

```python
# Find top 3 students who did most number of enrollments
regs.merge(students, on='student_id')[['student_id', 'name']].value_counts().head(3)
```

```python
# Find top 3 students who spend most amount of money on courses
idx = regs.merge(courses, on='course_id').groupby('student_id')['price'].sum().sort_values(ascending=False).head(3).index
students[students['student_id'].isin(idx)]['name'].values
```
