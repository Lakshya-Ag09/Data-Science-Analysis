```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# coloring on basis of categories
iris = pd.read_csv('Matplotlib_Dataset\\iris.csv')
print(iris.shape)
iris.head()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
iris.sample(5)
```

```python
plt.Figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# annotations
batters = pd.read_csv('Matplotlib_Dataset\\batter.csv')
batters.shape
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
sample_df
```

```python
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
# horizontal and vertical lines
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
plt.axhline(130, color='red')
plt.axvline(30, color='aqua')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

### Subplots

```python
batters.head()
```

```python
plt.scatter(batters['avg'], batters['strike_rate'])
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'], batters['strike_rate'])
ax.set_title('Title')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike rate')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1, sharex=True)

ax[0].scatter(batters['avg'], batters['strike_rate'])
ax[1].scatter(batters['avg'], batters['runs'])

ax[0].set_title('Avg vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].scatter(batters['avg'], batters['strike_rate'])
ax[0,1].scatter(batters['avg'], batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
fig = plt.figure(figsize=(8,10))
ax1 = fig.add_subplot(2,1,1)
ax1.scatter(batters['avg'], batters['strike_rate'])

ax2 = fig.add_subplot(2,1,2)
ax2.scatter(batters['avg'], batters['runs'])
```

### 3D Plots

```python
# 3d scatter plots
batters.head()
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter3D(batters['runs'], batters['avg'], batters['strike_rate'])
ax.set_title('Batsman analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('Strike Rate')
```

```python
# 3d line plot
x = [0,1,5]
y = [0,10,13]
z = [0,13,20]

fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter3D(x,y,z, s=[100,100,100])
ax.plot3D(x,y,z, color='red')
```

```python
# 3d surface plot
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

np.meshgrid(x,y)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
a = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(a)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
a = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(a)
```

```python
# contour plot
z = xx**2 + yy**2
fig = plt.figure()

ax = fig.add_subplot()
a = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(a)
```

```python
z = xx**2 + yy**2
fig = plt.figure()

ax = fig.add_subplot()
a = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(a)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()

ax = fig.add_subplot()
a = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(a)
```
