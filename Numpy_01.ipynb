### **Numpy Fundamentals**
Basic Numpy methods and operations

---

Numpy - Scientific Computing Library (Uses C prog in backend)
- Provide multidimensional array
- Fast operations
- Fixed size at creation

```python
# Importing Numpy
import numpy as np
```

Initializing 1D, 2D, 3D arrays using Numpy

```python
a = np.array([1,2,3])
print('a', a)

b = np.array([[1,2,3], [4,5,6]])
print('\nb', b)

c = np.array([[[1,2,3], [4,5,6]], [[1,2,3], [4,5,6]]])
print('\nc', c)
```

Changing datatype

```python
np.array([1,2,3], dtype=float)
```

Another way of Creating Numpy array using arange() function
- Similar to range() function
- arange(start, stop, skip)

```python
np.arange(1,12,2)
```

Changing shape/dimension of array using reshape() function

```python
# 2D array
np.arange(0,12).reshape(3,4)
```

```python
# 3D array
np.arange(0,8).reshape(2,2,2)
```

Different functions to create some special array

```python
np.ones((3,4))
```

```python
np.zeros((2,2))
```

```python
np.random.random((2,2))
```

```python
# Creating array where elements are linearly spaced
np.linspace(-9,9,10, dtype=int).reshape(2,5)
```

```python
np.identity(3)
```

Some functions of Numpy

```python
a1 = np.arange(10)
a2 = np.arange(9, dtype=np.int32).reshape(3,3)
a3 = np.arange(8, dtype=np.int8).reshape(2,2,2)
```

```python
# Checking dimensions of array
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
# Checking shape(Number of elements in each dimension)
print(a1.shape)
print(a2.shape)
print(a3.shape)
```

```python
# Checking total number of elements in array
print(a1.size)
print(a2.size)
print(a3.size)
```

```python
# Checking size in bytes occupied by each element
print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)
```

```python
# Checking datatype
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

Changing datatype of array

```python
# Here astype function created new array, previous array datatype doesn't get changed
a = a3.astype(np.int32)
print(a3.dtype)
print(a.dtype)
```

- Array functions

```python
print('a', a)
print('b', b)
```

```python
# max/min/sum/prod
np.sum(a)
```

```python
# axis 0=column, 1=row
np.max(a, axis=0)
```

```python
# mean/median/std/var
np.mean(a)
```

```python
np.median(a, axis=1)
```

```python
# Trigonometric Functions
np.sin(a)
```

```python
# Log and Exponents
print(np.log(a))
print(np.exp(b))
```

```python
# Round/Floor/Ceil
np.round(a/2)
```

Stacking

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12,24).reshape(3,4)
print(a,'\n\n',b)
```

```python
# Horizontal Stacking
np.hstack((a,b))
```

```python
np.hstack((a,b,b))
```

```python
# Vertical Stacking
np.vstack((a,b))
```

Splitting

```python
a = np.arange(24).reshape(4,6)
a
```

```python
# Horizontal Splitting
np.hsplit(a,2)
```

```python
np.hsplit(a,3)
```

```python
# Vertical Splitting
np.vsplit(a,2)
```
