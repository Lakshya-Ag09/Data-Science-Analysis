### **Working with Textual Data**


### Objective 
- We have dataset with reviews and their sentiment, either it is a positive review or negative review
- This is a NLP problem, where our purpose is to make a model which can predict new reviews sentiment

### Steps to perform:
1. Data gathering
2. Data cleaning
    - lowercasing
    - removing leading and trailing spaces
    - removing html tags
    - removing urls
    - expanding abbreviations (like, there's -> there is)
    - spelling correction
    - removing punctuations
    - remove special characters
3. Data preprocessing
    - tokenization (like converting sentence into list of every word)
    - stopword removal (words that doesn't help in sentiment meaning but used for sentence formation, eg: I, and, have, could, ...)
    - stemming (converting words into their root word, eg: dance, danced, dancing coverted to dance)
4. EDA (Exploratory Data Analysis)
5. Make features (eg: #words (number of words))
6. Vectorization (converting all text data into numbers)
    - BoW
    - TfIdf
    - word2vec
7. Modelling
8. Evaluation
9. Deployement
10. Monitoring

### What we'll perform in this file:
1. Data gathering
2. Data cleaning
3. Preprocessing
4. EDA
5. Make features
6. Vectorization

```python
# spelling correction
def spelling_correction(text):
    return TextBlob(text).correct().string
```

```python
df['review'] = df['review'].apply(spelling_correction)
df.head()
```

```python
# removing punctuations
string.punctuation 
```

```python
def remove_punctuation(text):
    for i in string.punctuation:
        if i in text:
            text = text.replace(i,'')
    
    return text
```

```python
df['review'] = df['review'].apply(remove_punctuation)
df.head()
```

```python
# removing special characters
df['review'] = df['review'].str.replace(r'[^\w\s]','', regex=True)
df.head()
```

**Preprocessing**

```python
# tokenization
df['tokenized_review'] = df['review'].apply(word_tokenize)
df.head()
```

```python
# stopword removal
stopwords.words('english')
```

```python
def remove_stopwords(text):

    L = []
    for word in text:
        if word not in stopwords.words('english'):
            L.append(word)
    
    return L
```

```python
df['tokenized_review'] = df['tokenized_review'].apply(remove_stopwords)
df.head()
```

**Vectorization**
- Bag of Words (BoW)
    - It first make vocabulary (list of all unique words)
    - Count the frequency of every unique word in the data and most frequent unique words will be captured
    - Then it makes vector representation of unique words that are captured and reviews (matrix of unique words captured, in every review)

```python
# Unigrams BoW
count_vectorizer = CountVectorizer(max_features=5000)
bag_of_words = count_vectorizer.fit_transform(df['review'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(), columns=count_vectorizer.get_feature_names_out())
print(bag_of_words.shape)
bag_of_words.head()
```

```python
# bigrams BoW
count_vectorizer = CountVectorizer(max_features=5000, ngram_range=(2,2))
bag_of_words = count_vectorizer.fit_transform(df['review'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(), columns=count_vectorizer.get_feature_names_out())
print(bag_of_words.shape)
bag_of_words.head()
```

```python
# trigrams BoW
count_vectorizer = CountVectorizer(max_features=5000, ngram_range=(3,3))
bag_of_words = count_vectorizer.fit_transform(df['review'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(), columns=count_vectorizer.get_feature_names_out())
print(bag_of_words.shape)
bag_of_words.head()
```

```python
# Best ngrams BoW
count_vectorizer = CountVectorizer(max_features=5000, ngram_range=(1,3))
bag_of_words = count_vectorizer.fit_transform(df['review'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(), columns=count_vectorizer.get_feature_names_out())
print(bag_of_words.shape)
bag_of_words.head()
```

### PCA
- used to convert high dimensional data into low dimensional data

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(bag_of_words.values)
print(pca_result.shape)
pca_result
```
